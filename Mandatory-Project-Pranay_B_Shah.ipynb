### Please explain which feature is nominal, ordinal, discrete or continuous
1. **Nominal** - These represent categories without any order, such as geographical coordinates (Latitude and Longitude), These are nominal as they represent locations on the map. They don't have a numerical order; they're just different points on the map.
2. **Ordinal** - These have a specific order but might not have a consistent difference between values.
3. **Continous** - These can take on any value within a range and often include fractional or decimal values. (Median House Value, Median Income, Median Age), which varies across various values or a range  
4. **Discrete** - These represent whole, distinct, and separate values, usually integers.(Population, Number of rooms and bedrooms)

## <font color='yellow'>  1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.</font>

- There are outliers which are after 10.
- The x-axis represents the range of median incomes
- The y-axis shows the number of count of each range. 
- The KDE - Kernel Density Estimate shows skewness, the plot is skewed to the right, it indicates that most of the data is concentrated towards one end of the income range or The mean is greater than the median, A right skewed distribution is also known as a positive skew
- KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate

## <font color='yellow'>10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.</font>

```python
def bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'small'
    if 11 <= total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

# create the new column 'total_bedroom_size' using the function
housing_fill['total_bedroom_size'] = housing_fill['total_bedrooms'].apply(bedroom_size)

# exporting new dataset with new column 
housing_fill.to_excel("Bedroom_size_housing_fill.xlsx")

# display the first 20 rows of the dataset with the new column
print(housing_fill[['total_bedrooms', 'total_bedroom_size']].head(20))
```
